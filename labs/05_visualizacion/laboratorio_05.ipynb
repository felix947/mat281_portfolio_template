1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns 
#grafico lineal
palette = sns.color_palette("hls", 6)
sns.lineplot(
    x='month_number',
    y='total_profit',
    data=df,
    ci = None,
    palette=palette
)   
```

```python
#grafico de dispersion
sns.scatterplot(
    x='month_number',
    y='total_profit',
    data=df,
    palette=palette
)   
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
#saque total_profit porque se veia mal con esos datos porque son numeros muy grandes
for i in df.columns:
    if i!='total_profit':
        sns.lineplot(
        x='month_number',
        y=i,
        data=df,
        ci = None,
        palette=palette
        )   
    
```
